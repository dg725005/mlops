```python
X = df[['Daytime/evening attendance','Previous qualification (grade)']]

y = df[['Target']]
```

```python
set(y.Target)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=63,test_size=0.15)
```

```python
y = pd.get_dummies(y,columns = ['Target']).replace({True:1.0, False:0.0})
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

classifier = RandomForestClassifier(n_estimators=80,max_depth=10,criterion='entropy')

model = classifier.fit(X_train, y_train)

print(accuracy_score(y_test,model.predict(X_test))*100)
```
